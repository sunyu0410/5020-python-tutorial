## Introduction to Plotting in Python

* `Matplotlib` is the most commonly used.
* `Seaborn` is a library built on top of matplotlib.
* Others such as `plotly`.

## Typical Workflow

In `matplotlib`, you deal with two types of objects:
  * **Figure**: the overall plot, e.g. set the size of the plot, the main title, can be set on the figure.
  * **Axes**: the content where data are drawn. This is like the canvas of a painting, where the objects get painted on.


You can either:
 * deal with figures and axes directly, or
 * use the `pyplot` submodule to handle them, which is much quicker and easier.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some data

np.random.seed(123)

x = np.random.random(100) * 10 - 5
x.sort()
noise = np.random.normal(0, 3, 100)
y = np.exp(x) + noise
```

```python
# Basic plot
plt.plot(x, y)
plt.show() # This line can be left out if run in a notebook
```

```python
# Styles
plt.plot(x, y, '-o')
plt.show()
```

```python
# Styles
plt.plot(x, y, '--')
plt.show()
```

> Marker styles can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle.filled_markers).

```python
plt.plot?
```

```python
# Styles
plt.plot(x, y, '-o', alpha=0.3, c='Orange')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Styles
plt.style.use('seaborn')
plt.plot(x, y, '-o', alpha=0.3, c='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Relationship between input and output')
plt.show()
```

> Plot styles can be found [here](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

```python
# Using figures and axes
fig, ax = plt.subplots(1,1)
ax.plot(x, y, '-o', alpha=0.3, c='red')
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.set_title('Relationship between input and output')
fig.show()
```

```python
# Saving
plt.plot(x, y, '-o', alpha=0.3, c='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Relationship between input and output')
plt.savefig('plot.png')
plt.clf()
```

## Seaborn

Let's do a quick one in `seaborn` for comparison. I think it's slightly nicer looking.

```python
import seaborn as sns
sns.set_theme()

sns.lineplot(x=x, y=y, alpha=0.8, linewidth=2)
sns.scatterplot(x=x, y=y, s=90, c='r', alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relationship between input and output')
plt.show()
```
